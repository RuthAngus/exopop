```python
import emcee
import numpy as np
import pandas as pd
import matplotlib.pyplot as pl
```

We'll start by computing the mean transit probability (at a period of 10 days) for this sample of stars.

```python
stars = pd.read_hdf("data/stlr.h5", "stlr")
G = 2945.4625385377644
P = 10.0
transit_lnprobs = (np.log(4*np.pi*np.pi) - np.log(G*P*P*stars.Mstar)) / 3
transit_lnprob0 = np.mean(transit_lnprobs)
print("The mean geometric transit probability at a 10 day period is {0}".format(np.exp(transit_lnprob0)))
```

Then, we'll load the completeness simulations.

```python
sim = pd.read_hdf("data/mcDV.h5", "mcDV")
m = sim.found * sim.bDV
log_per, log_rp = np.log(sim.inj_P), np.log(sim.inj_Rp)
censor = CensoringFunction(log_per, log_rp, m,
                           log_per_range=np.log([5, 400]), log_rp_range=np.log([0.5, 16]),
                           transit_lnprob0=transit_lnprob0, ln_period0=np.log(10.0))
```

And the planet candidates.

```python
log_per_obs = []
log_rp_obs = []
for line in open("data/table_ekoi836.tex"):
    cols = line.split("&")
    log_per_obs.append(np.log(float(cols[2])))
    log_rp_obs.append(np.log(float(cols[12])))

log_per_obs = np.array(log_per_obs, dtype=float)
log_rp_obs = np.array(log_rp_obs, dtype=float)
dataset = Dataset(log_per_obs, log_rp_obs, censor)
```

Now, here's our version of Figure 1 from Erik's paper:

```python
pl.pcolor(censor.log_per_bins, censor.log_rp_bins, np.exp(censor.lncompleteness).T, cmap="gray")
pl.plot(dataset.log_per_obs, dataset.log_rp_obs, ".r", ms=3)
pl.xlim(censor.log_per_bins.min(), censor.log_per_bins.max())
pl.ylim(censor.log_rp_bins.min(), censor.log_rp_bins.max())
pl.xlabel(r"$\ln P$")
pl.ylabel(r"$\ln R_P$")
pl.colorbar();
```

### A simple seperable population model

We'll start by modeling the population as separable. The period distribution is a broken power law and the radius distribution is a normalized histogram.

```python
log_per_dist = BrokenPowerLaw(censor.log_per_bins)
log_rp_dist = Histogram(censor.log_rp_bins, resample=8)
population0 = SeparablePopulation(log_per_dist, log_rp_dist)
```

And here's the probabilistic model that knows how to compute the ln-likelihood of the data given a set of model parameters:

```python
model0 = ProbabilisticModel(dataset, population0)
```

The population will suggest (terrible) inital parameters (flat in $\ln P$ and $\ln R_P$) and we can compute the ln-probability for these values:

```python
p0 = population0.initial()
print("The initial ln-probability is: {0}".format(model0.lnprob(p0)))
```

Now let's use emcee to sample this posterior probability density.

```python
ndim, nwalkers = len(p0), 32
pos = [p0 + 1e-8 * np.random.randn(ndim) for i in range(nwalkers)]

# Make sure that all the initial positions have finite probability.
assert np.all(np.isfinite(map(model0.lnprob, pos)))

# Set up the sampler.
sampler = emcee.EnsembleSampler(nwalkers, ndim, model0)
```

```python
pos, lp, state = sampler.run_mcmc(pos, 4000)
```

Here are 24 posterior samples for the period and radius distribution:

```python
samples = sampler.chain[:, 3000:, :]
samples = samples.reshape((-1, ndim))
population0.plot(samples[np.random.randint(len(samples), size=24)], title="posterior sampling");
```

### Including uncertainties

To include uncertainties, we use the `PHist`. The basic idea is that the catalog gives us samples from the posterior PDF $p(R\,|\,\hat{R},\sigma^2)$ for the true radius $R$ given the observed radius $\hat{R}$ and uncertainty $\sigma^2$. Combining this with our model of $p(P, R\,|\,\theta)$ from above, we can compute the "marginalized likelihood":

$$p(\hat{R},P\,|\,\theta,\sigma^2) = \int \mathrm{d}R \, p(R,\hat{R},P\,|\,\theta,\sigma^2)$$
$$= \int \mathrm{d}R \, p(R,P\,|\,\theta)\,p(\hat{R}\,|\,R,\sigma^2)$$
$$= \int \mathrm{d}R \, p(R,P\,|\,\theta)\,p(\hat{R}\,|\,R,\sigma^2)\,\frac{p(R\,|\,\hat{R},\sigma^2)}{p(R\,|\,\hat{R},\sigma^2)}$$
$$= Z\,\int \mathrm{d}R \, \frac{p(R,P\,|\,\theta)}{p(R)}\,p(R\,|\,\hat{R},\sigma^2)$$
$$\approx C \, \sum_{n=1}^N \frac{p(R^{(n)},P\,|\,\theta)}{p(R^{(n)})}$$

where the $R^{(n)}$ are samples from the posterior. We'll assume that Erik's radii were measured with uniform priors in radius here but the generalization would not be difficult.

```python
N = 24
noisy_log_per_obs = []
noisy_log_rp_obs = []
for line in open("data/table_ekoi836.tex"):
    cols = line.split("&")
    noisy_log_per_obs.append(np.log(float(cols[2])))
    noisy_log_rp_obs.append(np.log(float(cols[12]) + float(cols[13])*np.random.randn(N)))

noisy_log_per_obs = np.array(noisy_log_per_obs, dtype=float)
noisy_log_rp_obs = np.array(noisy_log_rp_obs, dtype=float)
noisy_dataset = Dataset(noisy_log_per_obs, noisy_log_rp_obs, censor)
```

This is what Figure 1 looks like now:

```python
pl.pcolor(censor.log_per_bins, censor.log_rp_bins, np.exp(censor.lncompleteness).T, cmap="gray")
pl.plot(noisy_dataset.log_per_obs, noisy_dataset.log_rp_obs, ".r", ms=2, alpha=0.3)
pl.xlim(censor.log_per_bins.min(), censor.log_per_bins.max())
pl.ylim(censor.log_rp_bins.min(), censor.log_rp_bins.max())
pl.xlabel(r"$\ln P$")
pl.ylabel(r"$\ln R_P$")
pl.colorbar();
```
